```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a=np.array([1,2,3,4,5])
b=np.array([6,7,8,9,1])
plt.scatter(a,b,)
```

```python
plt.scatter(a,b,s=40,color='m')
```

```python
a1=np.array([11,12,13,14,15])
b1=np.array([16,17,18,19,21])
plt.scatter(a1,b1,alpha=0.5) #alpha for transperancy
plt.scatter(a,b,s=40,color='m',marker="*")
```

```python
plt.bar(a,b,color='k',width=0.5)
```

```python
plt.barh(a,b,color="m",height=0.5)
```

```python
plt.hist(a)#create histogram for individual colum
```
